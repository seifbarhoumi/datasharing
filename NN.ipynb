## I. Approximation of a sinus function ##

The final goal of this question is to create a simple neural network model which is able to approximate a series of points create with a random generator and their corresponding sinus value.




```
import numpy as np

def randSinVector(n):
    np.random.seed(42) # to get same random values eveytimes
    x = np.empty([n,2])
    theta = 2 * np.pi * np.random.random_sample((n))
    x[:,0] = theta
    x[:,1] = np.sin(theta)  # value in radian
    return x
```
